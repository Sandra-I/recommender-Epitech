```python
def frequence_item(x):
    total_achat = x["TICKET_ID"].count()
    freq = x["LIBELLE"].value_counts() / total_achat 
    return freq
```

```python
z =df.groupby("CLI_ID").apply(frequence_item)
z = z.reset_index()
z = z.set_index("CLI_ID")
z.columns

```

```python
z
```

```python
user_item_frequence_pivot = z.pivot_table(index="CLI_ID",columns="level_1", values="LIBELLE")
user_item_frequence_pivot
```

```python
avg_user_item_freq = user_item_frequence_pivot.mean(axis=1)
user_item_frequence_pivot = user_item_frequence_pivot.sub(avg_user_item_freq, axis=0)
user_item_frequence_pivot = user_item_frequence_pivot.fillna(0)
```

```python
item_user_frequence_pivot = user_item_frequence_pivot.T
item_user_frequence_pivot
```

```python
q = cosine_similarity(item_user_frequence_pivot.loc['GD JDM4 PAMPLEMOUSSE FL 200ML', :].values.reshape(1, -1), item_user_frequence_pivot.loc['CR JR PARF BIO.SPE AC.SENT.50ML', :].values.reshape(1, -1))
q[0][0]
```

```python
similarities = cosine_similarity(item_user_frequence_pivot)
cosine_similarity_df = pd.DataFrame(similarities, index=item_user_frequence_pivot.index,columns=item_user_frequence_pivot.index)
cosine_similarity_df
```

```python
cosine_similarity_series = cosine_similarity_df.loc['GD JDM4 PAMPLEMOUSSE FL 200ML']
ordered_similarities = cosine_similarity_series.sort_values(ascending=False)
print(ordered_similarities)
```
