```python
df_cli_item = df[["CLI_ID", "LIBELLE"]]
```

```python
def create_pairs(x):
    pairs = pd.DataFrame(list(permutations(x.values,2)), columns=["item_a", "item_b"])
    return pairs
```

```python
df_pairs = df_cli_item.groupby("CLI_ID")["LIBELLE"].apply(create_pairs)
print(df_pairs.head())
```

```python
df_pairs
```

```python
df_cli_item2 = df_cli_item.drop_duplicates(subset=["CLI_ID","LIBELLE"], keep='first')
```

```python
df_cli_item2
```

```python
df_pairs2 = df_cli_item2.groupby("CLI_ID")["LIBELLE"].apply(create_pairs)
```

```python
df_pairs2 = df_pairs2.reset_index(drop=True)
```
